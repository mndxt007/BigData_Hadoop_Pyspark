```python
#Spark Imports and Get Context

#https://spark.apache.org/docs/latest/sql-getting-started.html
from pyspark import SparkContext, SparkConf, SQLContext
from os import getcwd

conf = SparkConf().setAppName('SparkS1')
sc = SparkContext(conf=conf).getOrCreate()
spark = SQLContext.getOrCreate(sc)
```

```python
#Exercise for Working with text file - Find the most used word
#Ref file - http://www.gutenberg.org/ebooks/1342

inputfileRDD = sc.textFile('../datasets/Pride_Prejudice.txt') #Creates RDD
#inputfileRDD.take(5) - Returns first 5 elements
inputfileRDD1 = inputfileRDD.filter(lambda x : x.strip() != '').map(lambda x: x.lower()) #Removes all lines which doesn't have any words and makes all characters lowercase.
#inputfileRDD1.take(5)
wordsRDD= inputfileRDD1.flatMap(lambda x: x.split(' ')) # Splits line to words and merges the arrays
wordsRDD.take(5)
```

```python
#Adding Value to each word for using it later for Aggregation

wordsRDDpair= wordsRDD.map(lambda x: (x,1))
#wordsRDDpair.take(5)

wordCount = wordsRDDpair.reduceByKey(lambda x,y : x+y)
#wordCount.take(5)

#Sorting based on occurance

wordCountSorted = wordCount.sortBy(lambda x : -x[1])
wordCountSorted.take(5)

#Save file
wordCountSorted.saveAsTextFile('../datasets/out/2_outfile.txt')
```
